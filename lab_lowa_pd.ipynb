```python
max_count = 0
the_store_number = None

for store_number in df["Store Number"].unique():
    count = sum(df["Store Number"] == store_number)

    if count > max_count:
        max_count = count
        the_store_number = store_number

print(
    f"The store number with the most data is {the_store_number} with {max_count} sales"
)
```

```python
store2count = dict()

for store_number in df["Store Number"].unique():
    count = sum(df["Store Number"] == store_number)

    if count >= 90:
        store2count[store_number] = count

print(store2count)
```
